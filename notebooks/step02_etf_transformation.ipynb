```python
import pandas as pd
import json
```

```python
# read in data for etf that targets the Nasdaq 100
etf_qqq = pd.read_csv("hist/QQQ.csv")

# S&P 500
etf_splg = pd.read_csv("hist/SPLG.csv")
etf_ivv = pd.read_csv("hist/IVV.csv")
etf_voo = pd.read_csv("hist/VOO.csv")
etf_spy = pd.read_csv("hist/SPY.csv")

# Dow Jones Industrial Average
etf_dia = pd.read_csv("hist/DIA.csv")

# Global
etf_acwi = pd.read_csv("hist/ACWI.csv")
etf_veu = pd.read_csv("hist/VEU.csv")
etf_vss = pd.read_csv("hist/VSS.csv")
```

```python
# add column for filtering data later
etf_qqq['symbol'] = 'QQQ'

etf_splg['symbol'] = 'SPLG'
etf_ivv['symbol'] = 'IVV'
etf_voo['symbol'] = 'VOO'
etf_spy['symbol'] = 'SPY'

etf_dia['symbol'] = 'DIA'

etf_acwi['symbol'] = 'ACWI'
etf_veu['symbol'] = 'VEU'
etf_vss['symbol'] = 'VSS'
```

```python
# concatenating the new DataFrames
etf_df = pd.concat([etf_qqq, etf_splg, etf_ivv, etf_voo, etf_spy, etf_dia, etf_acwi, etf_veu, etf_vss])
```

```python
etf_df
```

```python
# operations for converting a dataframe to a list of dictionaries

# use .to_dict and 'records' orient to make our desired list
list_d = etf_df.to_dict(orient='records')

# write to a text file for our data.js later
with open("etf_df.txt", "w") as file:
    file.write(json.dumps(list_d))
```
